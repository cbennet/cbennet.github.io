```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

dfTot = pd.read_csv("estimated_crimes_1979_2020.csv")
dfTot['state_abbr'].replace(NaN, "National_Total",inplace=True)
dfTot['state_name'].replace(NaN, "National_Total",inplace=True)


dfTot.head(5)
```

```python
crimeList = dfTot.columns[4:14]

dfTot["TotalCrime"] = dfTot[crimeList].sum(axis=1)
dfTot.head(5)
```

```python
for crime in crimeList:
  dfTot[crime+"_crimePercent"] = dfTot[crime] /dfTot["TotalCrime"]
dfTot.head(5)
```

```python
for crime in crimeList:
  dfTot[crime+"_popPercent"] = dfTot[crime] /dfTot["population"]
dfTot.head(5)
```

```python
dfAK = dfTot[dfTot["state_name"] == "Alaska"]
dfAK.head(5)
```

```python
for crime in crimeList:
  dfAK.plot.bar(x="year", y=crime+"_popPercent", title="Alaska "+ crime + " commited per population")
```

```python
states = dfTot["state_abbr"].unique()
print(states)
```

```python
for state in states:
  dfTmp = dfTot[dfTot["state_abbr"] == state]
  for crime in crimeList:
    dfTmp.plot.bar(x="year", y=crime+"_popPercent", subplots=True, title=state + " " + crime + " commited per population")
    dfTmp.plot.bar(x="year", y=crime+"_crimePercent", subplots=True, title=state + " " + crime + " commited per total crime")
```
